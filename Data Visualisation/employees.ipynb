```python
import pandas as pd
import matplotlib as plt

data = pd.read_csv("employees.csv")
data.loc[:,["sex", "age"]].groupby(by='sex').mean().plot(kind='bar', ylabel="Percentage")
```

```python
#data.loc[:,["sex", "salary"]].groupby(by='sex').plot(kind="hist")
data.loc[:,['salary','sex']].plot(kind="hist",title="Salary", ylabel="People", subplots=True, by='sex')
```

```python
data = pd.DataFrame(pd.read_csv("european-countries.csv"))
data.index = data["Name"]
data.loc[:,["Region","Name","Population"]].query('Region=="Central Europe"').sort_values(by='Population', ascending=False).plot(kind="bar",style="plain")
```

```python
data = pd.DataFrame(pd.read_csv("european-countries.csv"))
data.loc[:,["Region","Area"]].groupby(by="Region").sum().plot(kind="pie", subplots=True, autopct="%1.2f%%")
```
